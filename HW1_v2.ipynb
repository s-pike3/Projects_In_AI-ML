### 1.
 Since logistic regression has the form $\log\frac{p(x)}{1-p(x)} = \beta_0 + x * \beta$ \\
 $\frac{p(x)}{1-p(x)} = e^{\beta_0 + x * \beta}$ \\
  $p(x) = (1- p(x)) * e^{\beta_0 + x * \beta}$ \\
  $p(x) =  e^{\beta_0 + x * \beta} - p(x) * e^{\beta_0 + x * \beta}$ \\
  $p(x) +  p(x) * e^{\beta_0 + x * \beta} =  e^{\beta_0 + x * \beta}$ \\
  $p(x)= \frac{ e^{\beta_0 + x * \beta}}{1 + e^{\beta_0 + x * \beta}} = \frac{ 1}{1 + e^{-(\beta_0 + x * \beta)}}$


Using MLE, we want to maximize the probability of the data X, given our parameters $\beta$. For a given data point $x_i$, its class $y_i = 1$ with probability p and $y_i = 0$ with probability 1-p. Thus we want to maximize \begin{equation} \prod_{i \space in  \space y_i = 1} p(x_{i}) \end{equation} for all samples where $y_i =  1$ and
 \begin{equation} \prod_{i \space in  \space y_i = 0} (1 - p(x_{i})) \end{equation} for all samples where $y_i =  0$.

We can write the loss function as
\begin{equation}
L = \prod_{i \space in  \space y_i = 1} p(x_{i}) * \prod_{i \space in  \space y_i = 0} (1 - p(x_{i}))
\end{equation}
Combining the products,
\begin{equation}
 L = \prod_{i} (p(x_{i})^{y_i} * (1 - p(x_{i}))^{1 - y_i})
 \end{equation}
Converting to log-likelihood
\begin{equation}
l = \sum_{i=1}^{n}  y_{i}\log(p(x_i)) + (1 - y_{i})\log(1 - p(x_i))
\end{equation}
From the derivation above we see that $p(x)  = \frac{ 1}{1 + e^{-(\beta_0 + x * \beta)}}$. Thus we subsitute for $p(x)$,
\begin{equation}
l = \sum_{i=1}^{n}  y_{i}\log(\frac{ 1}{1 + e^{-(\beta_0 + x_i * \beta)}}) + (1 - y_{i}) \log(1 - \frac{ 1}{1 + e^{-(\beta_0 + x * \beta)}})
\end{equation}
Simplifying further,
\begin{equation}
l = \sum_{i=1}^{n}  y_{i}\log(\frac{ 1}{1 + e^{-(\beta_0 + x * \beta)}}) + (1 - y_{i}) \log( \frac{ e^{-(\beta_0 + x * \beta)}}{1 + e^{-(\beta_0 + x * \beta)}})
\end{equation}
\begin{equation}
l = \sum_{i=1}^{n}  y_{i}(\log(\frac{ 1}{1 + e^{-(\beta_0 + x * \beta)}}) - \log( \frac{ e^{-(\beta_0 + x * \beta)}}{1 + e^{-(\beta_0 + x * \beta)}})) + \log( \frac{ e^{-(\beta_0 + x * \beta)}}{1 + e^{-(\beta_0 + x * \beta)}})
\end{equation}
\begin{equation}
l = \sum_{i=1}^{n}  y_{i} * \log(  e^{(\beta_0 + x * \beta)}) + \log( \frac{ e^{-(\beta_0 + x * \beta)}}{1 + e^{-(\beta_0 + x * \beta)}} * \frac{e^{(\beta_0 + x * \beta)}}{e^{(\beta_0 + x * \beta)}})
\end{equation}
\begin{equation} l = \sum_{i=1}^{n}  y_{i} * (\beta_0 + x * \beta)) + \log( \frac{1}{1 + e^{(\beta_0 + x * \beta)}})
\end{equation}

\begin{equation} l = \sum_{i=1}^{n}  y_{i} * (\beta_0 + x * \beta)) - \log( 1 + e^{(\beta_0 + x * \beta)})
\end{equation}


#### MAP
In MLE, we find the optimal parameters that maximize the probability of the observed data. It can thus be formulated as $w^* = argmax_{W} P(D|W)$

In MAP, we use both the observed data and prior belief to find the optimal parameters. It is formulated as $w^* = argmax_{W} P(W|D)$. \\
Using Bayes, \\
$w^* = argmax_{W} P(W|D) = argmax_{W} \frac{P(D|W)P(W)}{P(D)}$ \\
Since $P(D)$ does not depend on $W$,
 $argmax_{w} \frac{P(D|W)P(W)}{P(D)} =  argmax_{W} P(D|W)P(W)$ \\
 Taking the log, \\
  $argmax_{w} \space \ln(P(D|W)) + \ln(P(W))$. \\
Thus MAP results in the same formulation as MLE with the additional term $\log(P(W))$.
\
According to Singh 2023, \begin{equation}
\ln(P(W)) \propto -\sum \frac{w_i^2}{2\kappa^2}
\end{equation}
This term penalizes large weight values. Thus MAP extends the formula from MLE by addding regularization.

```python
#compute metrics for model
def metrics(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1 = 2 * precision * recall / (precision + recall)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    print(f'Precision: {precision:.4f}')
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Accuracy: {accuracy:.4f}")
    return precision, recall, f1, accuracy
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
#Source: https://medium.com/@koushikkushal95/logistic-regression-from-scratch-dfb8527a4226
def get_loss(y_test, y_pred):
        epsilon = 1e-9 # add epsilon to preven NaNs
        #y1 = y_test * np.log(y_pred + epsilon)
        #y2 = (1-y_test) * np.log(1 - y_test + epsilon)
        return -np.mean(y_test * np.log(y_pred + epsilon) + (1-y_test) * np.log(1 - y_test + epsilon))
```

```python

def gd(X_train, y_train, num_features,epochs,lr):
  w = np.random.rand(num_features)
  loss = np.array([])
  for i in range(epochs):
    A = sigmoid(np.dot(X_train,w) )
    gw = 1/num_features * np.dot(X_train.T, A - y_train)
    w = w - lr * gw
    loss = np.append(loss, get_loss(y_train,A))
  return w, loss
```

```python
def sgd(X_train, y_train, num_features,epochs,lr):
  w = np.random.rand(num_features)
  loss = np.array([])
  for i in range(epochs):
    i = np.random.randint(0,len(X_train))
    A = sigmoid(np.dot(X_train[i],w) )
    gw = 1/num_features * np.dot(X_train[i], A - y_train[i])
    w = w - lr * gw
    A = sigmoid(np.dot(X_train,w) )
    loss = np.append(loss, get_loss(y_train,A))
  return w, loss
```

```python
def mini_batch_gd(X_train, y_train, num_features,epochs,lr,batch_size):
  w = np.random.rand(num_features)
  loss = np.array([])
  for i in range(epochs):
    idx = np.random.choice(np.arange(len(X_train)), batch_size, replace=False)
    x_b = X_train[idx]
    y_b = y_train[idx]
    A = sigmoid(np.dot(x_b,w) )
    gw = 1/num_features * np.dot(x_b.T, A - y_b)
    w = w - lr * gw
    A = sigmoid(np.dot(X_train,w) )
    loss = np.append(loss, get_loss(y_train,A))
  return w, loss
```

Batch Gradient Descent results

```python
w_gd, loss_gd = gd(X_train, y_train, 20,1000,0.001)
y_predicted = sigmoid(np.matmul(X_test,w_gd) )
y_predicted_cls = y_predicted.round()   # round off to nearest class
gd_results = metrics(y_test, y_predicted_cls)
```

Stochastic Gradient Descent results

```python
w_sgd, loss_sgd  = sgd(X_train, y_train, 20,500000,0.001)
y_predicted = sigmoid(np.matmul(X_test,w_sgd) )
y_predicted_cls = y_predicted.round()   # round off to nearest class
sgd_results = metrics(y_test, y_predicted_cls)
```

Mini-Batch Gradient Descent results

```python
w_mbgd, loss_mbgd  = mini_batch_gd(X_train, y_train, 20,50000,0.001,32)
y_predicted = sigmoid(np.matmul(X_test,w_mbgd) )
y_predicted_cls = y_predicted.round()   # round off to nearest class
mbgd_results = metrics(y_test, y_predicted_cls)
```

```python
plt.plot(np.arange(1, 1001), loss_gd, label="Batch Gradient Descent")
plt.plot(np.arange(1, 500001), loss_sgd, label="Stochastic Gradient Descent")
plt.plot(np.arange(1, 50001), loss_mbgd,label="Mini-Batch Gradient Descent")
plt.xscale('log')
plt.legend(loc='upper right')
plt.show()
```

Batch gradient descent converges after about 10 iterations, mini-batch gradient descent converges after about 1000 iterations, and stochastic gradient descent converges after about 500,000 epochs. Without computing the loss, it takes batch gradient descent about $2 * 10^{-3}$ seconds per epoch, stochastic gradient descent about $3 * 10^{-5}$ seconds per epoch, and mini-batch gradient descent about $5 * 10^{-3}$ per epoch. Thus, batch gradient descent was the most efficient.

Logistic Regression using PyTorch

```python
#Prepare data for training
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.astype(np.float32).reshape(-1,1)
y_test = y_test.astype(np.float32).reshape(-1,1)
```

```python
# Source : https://medium.com/@carlosrodrigo.coelho/logistic-regression-pytorch-956f96b28010
class LogisticRegression(nn.Module):

    def __init__(self, n_input_features=20):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(n_input_features, 1)

    def forward(self, x):
        y_predicted = torch.sigmoid(self.linear(x))
        return y_predicted

```

RMS Prop

```python
#Initialize PyTorch model with Skorch wrapper
model_rms = NeuralNetClassifier(
    module = LogisticRegression,
    criterion = nn.BCELoss,
    optimizer = torch.optim.RMSprop,
    optimizer__lr=0.05,
    optimizer__alpha=0.99

)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
param_grid = {
    'optimizer__lr': [0.01, 0.005, 0.001, 0.0005, 0.0001]
}
#Grid search for optimal learning rate
gs = GridSearchCV(estimator=model_rms, param_grid=param_grid, cv=3)
gs.fit(X_train, y_train)
y_pred = gs.predict(X_test)
```

```python
rms_results = metrics(y_test, y_pred)
```

#### Evaluation Metrics
I used precision, recall, F1 score, and accuracy to evaluate all of my models.

Accuracy can be measured by dividing the number of correct classifications by the total number of classifications. It measures the percentage that the model classifies correctly.

Precision is calculated by dividing the number of true positives by the number of predicted positives. It measures what percentage of the predicted positives (class 1) are true positives.

Recall is calculated by dividing the number of true positives by the number of positives in the test set. It measures the percentage of true positives (class 1) the model was able to correctly identify.

F1 is calculated as below: \\
$F1 = 2 * \frac{Precision * Recall}{Precision + Recall}$
It gives a balanced score of both precision and recall.
